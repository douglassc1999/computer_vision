```
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```
img = cv.imread('lontra.jpg', 0)
img = cv.resize(img, dsize=(500, 500), interpolation=cv.INTER_LINEAR)

cv2_imshow(img)
res = np.amax(img) - img
cv2_imshow(res)
```
