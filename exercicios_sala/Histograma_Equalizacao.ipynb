```
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```
img = cv.imread('lontra.jpg', 0)
imgcp = img
img = cv.resize(img, dsize=(500, 500), interpolation=cv.INTER_LINEAR)

cv2_imshow(img)

# Rk é a intensidade (entre 0 e L-1)
# Nk => número de ocorrências de Rk


```

```
# precisa estar ordenado
imgVec = img.flatten()
imgVec = np.sort(imgVec, axis = None) 

# Função que retorna a frequência da intensidade em um dicionario
def count_elements(seq) -> dict:
  hist = {}
  for i in seq:
    hist[i] = hist.get(i, 0) + 1
  return hist

# Ou podemos usar uma classe do Python para isso
from collections import Counter
recounted = Counter(imgVec)

counted = count_elements(imgVec)
counted == recounted

```

```
# usando matplotlib

# An "interface" to matplotlib.axes.Axes.hist() method
n, bins, patches = plt.hist(x=imgVec, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Intensidade')
plt.ylabel('Frequencia')
plt.title('Histograma')
# plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
# Set a clean upper y-axis limit.
# plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```
