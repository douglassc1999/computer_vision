```python
# importando módulo opencv
import cv2 as cv
print(cv.__version__)

# importando o numpy
import numpy as np
```

<h2>Getting Started with Images</h2>
Usa-se a função cv.imread() para ler uma imagem. A imagem deve estar no mesmo diretório de trabalho ou dever ser fornecido o caminho absoluto.

<h3>Flags</h3>
<ul>
    <li>cv.IMREAD_COLOR : Carrega uma imagem com cor. Qualquer transparência será negligenciada. É o padrão </li>
<li>cv.IMREAD_GRAYSCALE : Carrega numa escala de cinza</li>
<li>cv.IMREAD_UNCHANGED : Carrega a imagem sem alterações, incluindo o canal alpha(?) </li>
</ul>

```python
img = cv.imread('imagens/juninho.jpg', 0)
```

<h2>Display an image</h2>
Usa-se a função cv.imshow() para mostrar uma imagem. A janela vai se ajustar ao tamnho da imagem automaticamente.

```python
# cv.namedWindow('imagem', cv.WINDOW_NORMAL)
cv.imshow('imagem', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

<h2>Write an image</h2>
Usa-se a função cv.imwrite() para salvar uma imagem.
O primeiro argumento é o nome do arquico, o segundo é a imagem que se quer salvar

```python
cv.imwrite('imagens/juninhogray.png', img)
# salvou no caminho escolhido, dentro da pasta 'imagens' 
```

<h2>Using Matplotlib</h2>
O Matplotlib é uma biblioteca para plotagem com uma variedade de métodos. 

```python
# importando módulo utilizado
from matplotlib import pyplot as plt
```

```python
plt.imshow(img, cmap='gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([]) # para esconder eixos
plt.show()
```

<h2>Curiosity</h2>
O import com opencv de uma imagem vem no padrão BGR, enquanto o matplotlib segue o padrão RGB de cores. Então, tem-se que fazer uma conversão para o padrão RGB.

```python
img1 = cv.imread('imagens/juninho.jpg')

b,g,r = cv.split(img1)
img2 = cv.merge([r,g,b])

plt.subplot(121);plt.imshow(img1)# cores reais
plt.xticks([]), plt.yticks([])   # para esconder eixos
plt.subplot(122);plt.imshow(img2)# cores distorcidas
plt.xticks([]), plt.yticks([])   # para esconder eixos
plt.show()

cv.imshow('bgr image',img1) # cores reais
cv.imshow('rgb image',img2) # cores distorcidas
cv.waitKey(0)
cv.destroyAllWindows()
```
